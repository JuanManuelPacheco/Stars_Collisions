### 1) Determination of bound-unbound particles:

To achieve the central objective of this notebook the first thing to be done is the calculation of the total angular momentum of our remnant, i.e. the sum of all the angular momenta for the bounded particles after the collision, so the first thing to be done is the characterisation of the particles belonging to the remnant making use of all the machinery described in the **Menc_Munb.py** module.

```python
# Imported variables from b = 1 and Renzo collision simulation.

x, y, z, vx, vy, vz, m, d, u = np.genfromtxt('{}b1/out0882.sph.ascii'.format(data_path),
                                             usecols= (0,1,2,3,4,5,6,7,8), unpack=True)

xR, yR, zR, vxR, vyR, vzR, mR, dR, uR = np.genfromtxt('{}Renzo/out1606.sph.ascii'.format(data_path),
                                                      usecols= (0,1,2,3,4,5,6,7,8), unpack=True)
```

```python
tb1 = 1.8445e-02*0.5*882
tR = 1.8445e-02*0.5*1606
```

```python
nw_x, nw_y, nw_z, nw_vx, nw_vy, nw_vz = Men.re_center(x,y,z,vx,vy,vz,d)
nw_xR, nw_yR, nw_zR, nw_vxR, nw_vyR, nw_vzR = Men.re_center(xR,yR,zR,vxR,vyR,vzR,dR)
```

```python
X, Y, Z, VX, VY, VZ, R, V, M, U = Men.re_order(nw_x,nw_y,nw_z,nw_vx,nw_vy,nw_vz,m,u)
XR, YR, ZR, VXR, VYR, VZR, RR, VR, MR, UR = Men.re_order(nw_xR,nw_yR,nw_zR,nw_vxR,nw_vyR,nw_vzR,mR,uR)
```

```python
Mt, M_enc = Men.mass_quantities(M)
MtR, M_encR = Men.mass_quantities(MR)
```

```python
e = Men.energy(V,U,R,M_enc)
eR = Men.energy(VR,UR,RR,M_encR)
```

```python
bn, un = Men.bound_unbound(R,e)
bnR, unR = Men.bound_unbound(RR,eR)
```

### 2) Computation of L and aligment with z-axis:

The next step consists of calculating the total angular momentum of the remnant in each snapshot to be analyzed, with the subsequent objective of rotating the coordinate axis of all our particles in such a way that the z-axis coincides with this total angular momentum. El procedimiento matemático efectuado se encuentra plenamente descrito en cada una de las funciones documentadas en el módulo **L_evo.py**.

```python
pos_un, vel_un = np.array([X[un],Y[un],Z[un]]), np.array([VX[un],VY[un],VZ[un]])
pos_bn, vel_bn = np.array([X[bn],Y[bn],Z[bn]]), np.array([VX[bn],VY[bn],VZ[bn]])

posR_un, velR_un = np.array([XR[unR],YR[unR],ZR[unR]]), np.array([VXR[unR],VYR[unR],VZR[unR]])
posR_bn, velR_bn = np.array([XR[bnR],YR[bnR],ZR[bnR]]), np.array([VXR[bnR],VYR[bnR],VZR[bnR]])
```

```python
L_un, Lt_un = Lev.angular_momemtum(pos_un,vel_un,M[un])
L_bn, Lt_bn = Lev.angular_momemtum(pos_bn,vel_bn,M[bn])

LR_un, LtR_un = Lev.angular_momemtum(posR_un,velR_un,MR[unR])
LR_bn, LtR_bn = Lev.angular_momemtum(posR_bn,velR_bn,MR[bnR])
```

```python
pos_un_rot, vel_un_rot = Lev.rotation(Lt_bn,pos_un,vel_un)
pos_bn_rot, vel_bn_rot = Lev.rotation(Lt_bn,pos_bn,vel_bn)

posR_un_rot, velR_un_rot = Lev.rotation(LtR_bn,posR_un,velR_un)
posR_bn_rot, velR_bn_rot = Lev.rotation(LtR_bn,posR_bn,velR_bn)
```

```python
L_un_rot, Lt_un_rot = Lev.angular_momemtum(pos_un_rot,vel_un_rot,M[un])
L_bn_rot, Lt_bn_rot = Lev.angular_momemtum(pos_bn_rot,vel_bn_rot,M[bn])

LR_un_rot, LtR_un_rot = Lev.angular_momemtum(posR_un_rot,velR_un_rot,MR[unR])
LR_bn_rot, LtR_bn_rot = Lev.angular_momemtum(posR_bn_rot,velR_bn_rot,MR[bnR])
```

### 2) Radial, tangential and vertical velocity of the remnant:

Now that our coordinate system is aligned with the rotation axis of the remnant, we can calculate the radial, tangential and vertical velocities by performing a transformation from Cartesian to cylindrical coordinates. The idea is to be able to visualize each of the velocity components along the radius, and at the same time be able to characterize their temporal evolution throughout the development of the collision.

Given that the mean velocities showed that the only remnants with rotation are those from collisions with vertical separation between the stars, the standard deviation of the velocities is an excellent indicator that allows us to characterize collision instants much more accurately than the percentage of unbound mass. When all the deviations agree with an increasing peak, we can properly say that subsequent collisions are taking place, as we can identify in the quadrant representing the evolution of the simulation *b = 1*.

To further investigate the criterion of when to stop the simulations in the case where $b \neq 0$, we propose to use a criterion that allows us to compare the rotational motion of the remnant with the standard deviation ($\frac{V_i}{\sigma_i}$). If at the moment of the division between both quantities we can visualize a tendency to unity (as we can see in the graph below for *b = 0.1* and *b = 1*) we will know that the remnant is already in a state of sufficient relaxation.

```python
fig, axs=plt.subplots(nrows=2,  ncols=3, figsize=(15,10))
fig.subplots_adjust(hspace=0.3)

axs[0,0].set_title(r'b = 0.1', fontsize=22)
axs[0,0].plot(timeb0,Vrb0/stdrb0,'-g', label = r'$V_{r}/\sigma_{r}$')
axs[0,0].plot(timeb0,Vtb0/stdtb0,'-b', label = r'$V_{\theta}/\sigma_{\theta}$')
axs[0,0].plot(timeb0,Vzb0/stdzb0,'-r', label = r'$V_{z}/\sigma_{z}$')
axs[0,0].set_ylabel(r'$V$ [km $s^{-1}$]', fontsize = 20)
axs[0,0].legend(fontsize = 13, loc=0)
axs[0,0].grid()

axs[0,1].set_title(r'b = 1', fontsize=22)
axs[0,1].plot(timeb1,Vrb1/stdrb1,'-g', label = r'$V_{r}/\sigma_{r}$')
axs[0,1].plot(timeb1,Vtb1/stdtb1,'-b', label = r'$V_{\theta}/\sigma_{\theta}$')
axs[0,1].plot(timeb1,Vzb1/stdzb1,'-r', label = r'$V_{z}/\sigma_{z}$')
axs[0,1].legend(fontsize = 13, loc=0)
axs[0,1].grid()

axs[0,2].set_title(r'v = 100', fontsize=22)
axs[0,2].plot(timev1,Vrv1/stdrv1,'-g', label = r'$V_{r}/\sigma_{r}$')
axs[0,2].plot(timev1,Vtv1/stdtv1,'-b', label = r'$V_{\theta}/\sigma_{\theta}$')
axs[0,2].plot(timev1,Vzv1/stdzv1,'-r', label = r'$V_{z}/\sigma_{z}$')
axs[0,2].legend(fontsize = 13, loc=0)
axs[0,2].grid()

axs[1,0].set_title(r'v = 500', fontsize=22)
axs[1,0].plot(timev5,Vrv5/stdrv5,'-g', label = r'$V_{r}/\sigma_{r}$')
axs[1,0].plot(timev5,Vtv5/stdtv5,'-b', label = r'$V_{\theta}/\sigma_{\theta}$')
axs[1,0].plot(timev5,Vzv5/stdzv5,'-r', label = r'$V_{z}/\sigma_{z}$')
axs[1,0].set_ylabel(r'$V$ [km $s^{-1}$]', fontsize = 20)
axs[1,0].set_xlabel(r'$Time$ [Days]', fontsize = 20)
axs[1,0].legend(fontsize = 13, loc=0)
axs[1,0].grid()

axs[1,1].set_title(r'4 Myr', fontsize=22)
axs[1,1].plot(time4,Vr4/stdr4,'-g', label = r'$V_{r}/\sigma_{r}$')
axs[1,1].plot(time4,Vt4/stdt4,'-b', label = r'$V_{\theta}/\sigma_{\theta}$')
axs[1,1].plot(time4,Vz4/stdz4,'-r', label = r'$V_{z}/\sigma_{z}$')
axs[1,1].set_xlabel(r'$Time$ [Days]', fontsize = 20)
axs[1,1].legend(fontsize = 13, loc=0)
axs[1,1].grid()

axs[1,2].set_title(r'Renzo', fontsize=22)
axs[1,2].plot(timeR,VrR/stdrR,'-g', label = r'$V_{r}/\sigma_{r}$')
axs[1,2].plot(timeR,VtR/stdtR,'-b', label = r'$V_{\theta}/\sigma_{\theta}$')
axs[1,2].plot(timeR,VzR/stdzR,'-r', label = r'$V_{z}/\sigma_{z}$')
axs[1,2].set_xlabel(r'$Time$ [Days]', fontsize = 20)
axs[1,2].legend(fontsize = 13, loc=0)
axs[1,2].grid()

fig.tight_layout()
plt.show()
```
