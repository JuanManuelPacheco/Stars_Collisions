```python
def Phi_Gebbleek(c2,q):
    Phi = c2*q/((1+q)**2)
    return Phi

def Phi_Lombardi(c1,q,r1_86,r1_50,r2_86,r2_50):
    Phi = c1*(q/((1+q)**2))*((r1_86 + r2_86)/(r1_50 + r2_50))
    return Phi
```
