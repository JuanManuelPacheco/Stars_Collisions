```python
import numpy as np

import matplotlib as mpl
mpl.rcParams['agg.path.chunksize'] = 5000
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
from scipy.signal import find_peaks
```

```python
# Path where the ascii files of the snapshots are located.
data_path = '/home/juanmanuel/Documents/Master/Thesis/Simulations/'
```

```python
mfR,AfR,tfR,mR,AR,tR = np.genfromtxt('{}Renzo/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
mf4,Af4,tf4,m4,A4,t4 = np.genfromtxt('{}4Myr/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
mfH,AfH,tfH,mH,AH,tH = np.genfromtxt('{}Halfres/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
```

```python
aR = np.histogram(mR[2:], bins=150, weights=np.diff(np.diff(AR))*mR[2:])
bR = np.histogram(mR[2:], bins=150, weights=mR[2:])
cR = aR[0]/bR[0]
peaksR, _R = find_peaks(cR, height=0)
massR = aR[1][1:]

a4 = np.histogram(m4[2:], bins=150, weights=np.diff(np.diff(A4))*m4[2:])
b4 = np.histogram(m4[2:], bins=150, weights=m4[2:])
c4 = a4[0]/b4[0]
peaks4, _4 = find_peaks(c4, height=0)
mass4 = a4[1][1:]

aH = np.histogram(mH[2:], bins=150, weights=np.diff(np.diff(AH))*mH[2:])
bH = np.histogram(mH[2:], bins=150, weights=mH[2:])
cH = aH[0]/bH[0]
peaksH, _H = find_peaks(cH, height=0)
massH = aH[1][1:]
```

```python
mfb,Afb,tfb,mb,Ab,tb = np.genfromtxt('{}b0.1/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
mfb1,Afb1,tfb1,mb1,Ab1,tb1 = np.genfromtxt('{}b1/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
mfb3,Afb3,tfb3,mb3,Ab3,tb3 = np.genfromtxt('{}b3/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
```

```python
ab = np.histogram(mb[2:], bins=100, weights=np.diff(np.diff(Ab))*mb[2:])
bb = np.histogram(mb[2:], bins=100, weights=mb[2:])
cb = ab[0]/bb[0]
peaksb, _b = find_peaks(cb, height=0)
massb = ab[1][1:]

ab1 = np.histogram(mb1[2:], bins=100, weights=np.diff(np.diff(Ab1))*mb1[2:])
bb1 = np.histogram(mb1[2:], bins=100, weights=mb1[2:])
cb1 = ab1[0]/bb1[0]
peaksb1, _b1 = find_peaks(cb1, height=0)
massb1 = ab1[1][1:]

ab3 = np.histogram(mb3[2:], bins=100, weights=np.diff(np.diff(Ab3))*mb3[2:])
bb3 = np.histogram(mb3[2:], bins=100, weights=mb3[2:])
cb3 = ab3[0]/bb3[0]
peaksb3, _b3 = find_peaks(cb3, height=0)
massb3 = ab3[1][1:]
```

```python
mfv,Afv,tfv,mv,Av,tv = np.genfromtxt('{}v10/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
mfv1,Afv1,tfv1,mv1,Av1,tv1 = np.genfromtxt('{}v100/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
mfv5,Afv5,tfv5,mv5,Av5,tv5 = np.genfromtxt('{}v500/Entropy.dat'.format(data_path),dtype='float',
                       usecols=(0,1,2,3,4,5),skip_header=1,comments="#", unpack=True)
```

```python
av = np.histogram(mv[2:], bins=100, weights=np.diff(np.diff(Av))*mv[2:])
bv = np.histogram(mv[2:], bins=100, weights=mv[2:])
cv = av[0]/bv[0]
peaksv, _v = find_peaks(cv, height=0)
massv = av[1][1:]

av1 = np.histogram(mv1[2:], bins=100, weights=np.diff(np.diff(Av1))*mv1[2:])
bv1 = np.histogram(mv1[2:], bins=100, weights=mv1[2:])
cv1 = av1[0]/bv1[0]
peaksv1, _v1 = find_peaks(cv1, height=0)
massv1 = av1[1][1:]

av5 = np.histogram(mv5[2:], bins=100, weights=np.diff(np.diff(Av5))*mv5[2:])
bv5 = np.histogram(mv5[2:], bins=100, weights=mv5[2:])
cv5 = av5[0]/bv5[0]
peaksv5, _v5 = find_peaks(cv5, height=0)
massv5 = av5[1][1:]
```
