```python
from astropy import constants as cons
from astropy import units as un
import numpy as np
import matplotlib as mpl
mpl.rcParams['agg.path.chunksize'] = 100
import matplotlib.pyplot as plt
import sys
plt.rcParams.update({'font.size': 20})
```

```python
data_path = '/home/juanmanuel/Documents/Master/Thesis/Simulations/'
```

```python
VrR, VtR, VzR, stdrR, stdtR, stdzR = np.genfromtxt('{}Renzo/L_evo/L_evo.txt'.format(data_path),
                                                   usecols= (0,1,2,3,4,5), unpack=True)
VrH, VtH, VzH, stdrH, stdtH, stdzH = np.genfromtxt('{}Halfres/L_evo/z_evo.txt'.format(data_path),
                                                   usecols= (0,1,2,3,4,5), unpack=True)
Vr4, Vt4, Vz4, stdr4, stdt4, stdz4 = np.genfromtxt('{}4Myr/L_evo/L_evo.txt'.format(data_path),
                                                   usecols= (0,1,2,3,4,5), unpack=True)

Vrb, Vtb, Vzb, stdrb, stdtb, stdzb = np.genfromtxt('{}b0.1/L_evo/L_evo.txt'.format(data_path),
                                                         usecols= (0,1,2,3,4,5), unpack=True)
Vrb1, Vtb1, Vzb1, stdrb1, stdtb1, stdzb1 = np.genfromtxt('{}b1/L_evo/L_evo.txt'.format(data_path),
                                                         usecols= (0,1,2,3,4,5), unpack=True)
Vrb3, Vtb3, Vzb3, stdrb3, stdtb3, stdzb3 = np.genfromtxt('{}b3/L_evo/z_evo.txt'.format(data_path),
                                                         usecols= (0,1,2,3,4,5), unpack=True)

Vrv, Vtv, Vzv, stdrv, stdtv, stdzv = np.genfromtxt('{}v10/L_evo/z_evo.txt'.format(data_path),
                                                         usecols= (0,1,2,3,4,5), unpack=True)
Vrv1, Vtv1, Vzv1, stdrv1, stdtv1, stdzv1 = np.genfromtxt('{}v100/L_evo/L_evo.txt'.format(data_path),
                                                         usecols= (0,1,2,3,4,5), unpack=True)
Vrv5, Vtv5, Vzv5, stdrv5, stdtv5, stdzv5 = np.genfromtxt('{}v500/L_evo/L_evo.txt'.format(data_path),
                                                         usecols= (0,1,2,3,4,5), unpack=True)
```

```python
timeR = np.linspace(0,1.8445e-02*0.5*len(VrR),len(VrR))
timeH = np.linspace(0,1.8445e-02*0.5*len(VrH),len(VrH))
time4 = np.linspace(0,1.8445e-02*0.5*len(Vr4),len(Vr4))

timeb = np.linspace(0,1.8445e-02*0.5*len(Vrb),len(Vrb))
timeb1 = np.linspace(0,1.8445e-02*0.5*len(Vrb1),len(Vrb1))
timeb3 = np.linspace(0,1.8445e-02*0.5*len(Vrb3),len(Vrb3))                    
                    
timev = np.linspace(0,1.8445e-02*0.5*len(Vrv),len(Vrv))
timev1 = np.linspace(0,1.8445e-02*0.5*len(Vrv1),len(Vrv1))
timev5 = np.linspace(0,1.8445e-02*0.5*len(Vrv5),len(Vrv5))
```

```python
fig, axs=plt.subplots(nrows=3,  ncols=3, figsize=(22,18))
fig.subplots_adjust(hspace=0.3)

axs[0,0].set_title(r'Renzo, et al (2020)', fontsize=25)
axs[0,0].plot(timeR,VrR,'-k', linewidth=3., label = r'$V_{R}$')
axs[0,0].plot(timeR,VtR,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[0,0].plot(timeR,VzR,'-r', linewidth=3., label = r'$V_{z}$')
axs[0,0].set_ylabel(r'$V$ [km/s]', fontsize = 22)
axs[0,0].legend(fontsize = 22, loc=0)
axs[0,0].set_xlim([1.3,14.5])
axs[0,0].set_ylim([-20,60])

axs[0,1].set_title(r'Half Resolution', fontsize=25)
axs[0,1].plot(timeH,VrH,'-k', linewidth=3., label = r'$V_{R}$')
axs[0,1].plot(timeH,VtH,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[0,1].plot(timeH,VzH,'-r', linewidth=3., label = r'$V_{z}$')
axs[0,1].legend(fontsize = 22, loc=0)
axs[0,1].set_xlim([1.3,13.])
axs[0,1].set_ylim([-20,60])

axs[0,2].set_title(r'+2 $Myr$', fontsize=25)
axs[0,2].plot(time4,Vr4,'-k', linewidth=3., label = r'$V_{R}$')
axs[0,2].plot(time4,Vt4,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[0,2].plot(time4,Vz4,'-r', linewidth=3., label = r'$V_{z}$')
axs[0,2].legend(fontsize = 22, loc=0)
axs[0,2].set_xlim([1.3,4.3])
axs[0,2].set_ylim([-20,60])

axs[1,0].set_title(r'$b = 0.1$ $R_{\odot}$', fontsize=25)
axs[1,0].plot(timeb,Vrb,'-k', linewidth=3., label = r'$V_{R}$')
axs[1,0].plot(timeb,Vtb,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[1,0].plot(timeb,Vzb,'-r', linewidth=3., label = r'$V_{z}$')
axs[1,0].set_ylabel(r'$V$ [km/s]', fontsize = 22)
axs[1,0].legend(fontsize = 22, loc=0)
axs[1,0].set_xlim([1.3,8.3])
axs[1,0].set_ylim([-20,160])

axs[1,1].set_title(r'$b = 1$ $R_{\odot}$', fontsize=25)
axs[1,1].plot(timeb1,Vrb1,'-k', linewidth=3., label = r'$V_{R}$')
axs[1,1].plot(timeb1,Vtb1,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[1,1].plot(timeb1,Vzb1,'-r', linewidth=3., label = r'$V_{z}$')
axs[1,1].legend(fontsize = 22, loc=0)
axs[1,1].set_xlim([1.3,8.])
axs[1,1].set_ylim([-20,420])

axs[1,2].set_title(r'$b = 3$ $R_{\odot}$', fontsize=25)
axs[1,2].plot(timeb3,Vrb3,'-k', linewidth=3., label = r'$V_{R}$')
axs[1,2].plot(timeb3,Vtb3,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[1,2].plot(timeb3,Vzb3,'-r', linewidth=3., label = r'$V_{z}$')
axs[1,2].legend(fontsize = 22, loc=0)
axs[1,2].set_xlim([1.3,12.])
axs[1,2].set_ylim([-20,450])

axs[2,0].set_title(r'$v_{\infty} = 10$ $km/s$', fontsize=25)
axs[2,0].plot(timev,Vrv,'-k', linewidth=3., label = r'$V_{R}$')
axs[2,0].plot(timev,Vtv,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[2,0].plot(timev,Vzv,'-r', linewidth=3., label = r'$V_{z}$')
axs[2,0].set_ylabel(r'$V$ [km/s]', fontsize = 22)
axs[2,0].set_xlabel(r'$Time$ [days]', fontsize = 22)
axs[2,0].legend(fontsize = 22, loc=0)
axs[2,0].set_xlim([1.3,11.5])
axs[2,0].set_ylim([-30,60])

axs[2,1].set_title(r'$v_{\infty} = 100$ $km/s$', fontsize=25)
axs[2,1].plot(timev1,Vrv1,'-k', linewidth=3., label = r'$V_{R}$')
axs[2,1].plot(timev1,Vtv1,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[2,1].plot(timev1,Vzv1,'-r', linewidth=3., label = r'$V_{z}$')
axs[2,1].set_xlabel(r'$Time$ [days]', fontsize = 22)
axs[2,1].legend(fontsize = 22, loc=0)
axs[2,1].set_xlim([1.3,2.2])
axs[2,1].set_ylim([-20,60])

axs[2,2].set_title(r'$v_{\infty} = 500$ $km/s$', fontsize=25)
axs[2,2].plot(timev5,Vrv5,'-k', linewidth=3., label = r'$V_{R}$')
axs[2,2].plot(timev5,Vtv5,'-b', linewidth=3., label = r'$V_{\theta}$')
axs[2,2].plot(timev5,Vzv5,'-r', linewidth=3., label = r'$V_{z}$')
axs[2,2].set_xlabel(r'$Time$ [days]', fontsize = 22)
axs[2,2].legend(fontsize = 22, loc=0)
axs[2,2].set_xlim([1.3,11.5])
axs[2,2].set_ylim([-20,60])

fig.tight_layout()
plt.show()
```
