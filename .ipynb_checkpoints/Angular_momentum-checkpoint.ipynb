```python
# Needed libraries.
from astropy import units
import numpy as np
import matplotlib as mpl
mpl.rcParams['agg.path.chunksize'] = 10000
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
import sys
```

```python
sys.path.append("./Python/")
import Menc_Munb as Men
import L_evo as Lev
```

```python
# Path where the ascii files of the snapshots are located.
data_path = '/home/juanmanuel/Documents/Master/Semester2/Computational/'
```

```python
# Imported variables from b = 1 and Renzo collision simulation.

x, y, z, vx, vy, vz, m, d, u = np.genfromtxt('{}b1/out0882.sph.ascii'.format(data_path),
                                             usecols= (0,1,2,3,4,5,6,7,8), unpack=True)

xR, yR, zR, vxR, vyR, vzR, mR, dR, uR = np.genfromtxt('{}Renzo/out1606.sph.ascii'.format(data_path),
                                                      usecols= (0,1,2,3,4,5,6,7,8), unpack=True)
```

```python
nw_x, nw_y, nw_z, nw_vx, nw_vy, nw_vz = Men.re_center(x,y,z,vx,vy,vz,d)
nw_xR, nw_yR, nw_zR, nw_vxR, nw_vyR, nw_vzR = Men.re_center(xR,yR,zR,vxR,vyR,vzR,dR)
```

```python
X, Y, Z, VX, VY, VZ, R, V, M, U = Men.re_order(nw_x,nw_y,nw_z,nw_vx,nw_vy,nw_vz,m,u)
XR, YR, ZR, VXR, VYR, VZR, RR, VR, MR, UR = Men.re_order(nw_xR,nw_yR,nw_zR,nw_vxR,nw_vyR,nw_vzR,mR,uR)
```

```python
Mt, M_enc = Men.mass_quantities(M)
MtR, M_encR = Men.mass_quantities(MR)
```

```python
e = Men.energy(V,U,R,M_enc)
eR = Men.energy(VR,UR,RR,M_encR)
```

```python
bn, un = Men.bound_unbound(R,e)
bnR, unR = Men.bound_unbound(RR,eR)
```

```python
pos_un, vel_un = np.array([X[un],Y[un],Z[un]]), np.array([VX[un],VY[un],VZ[un]])
pos_bn, vel_bn = np.array([X[bn],Y[bn],Z[bn]]), np.array([VX[bn],VY[bn],VZ[bn]])

posR_un, velR_un = np.array([XR[unR],YR[unR],ZR[unR]]), np.array([VXR[unR],VYR[unR],VZR[unR]])
posR_bn, velR_bn = np.array([XR[bnR],YR[bnR],ZR[bnR]]), np.array([VXR[bnR],VYR[bnR],VZR[bnR]])
```

```python
L_un, Lt_un = Lev.angular_momemtum(pos_un,vel_un,M[un])
L_bn, Lt_bn = Lev.angular_momemtum(pos_bn,vel_bn,M[bn])

LR_un, LtR_un = Lev.angular_momemtum(posR_un,velR_un,MR[unR])
LR_bn, LtR_bn = Lev.angular_momemtum(posR_bn,velR_bn,MR[bnR])
```

```python
pos_un_rot, vel_un_rot = Lev.rotation(Lt_bn,pos_un,vel_un)
pos_bn_rot, vel_bn_rot = Lev.rotation(Lt_bn,pos_bn,vel_bn)

posR_un_rot, velR_un_rot = Lev.rotation(LtR_bn,posR_un,velR_un)
posR_bn_rot, velR_bn_rot = Lev.rotation(LtR_bn,posR_bn,velR_bn)
```

```python
L_un_rot, Lt_un_rot = Lev.angular_momemtum(pos_un_rot,vel_un_rot,M[un])
L_bn_rot, Lt_bn_rot = Lev.angular_momemtum(pos_bn_rot,vel_bn_rot,M[bn])

LR_un_rot, LtR_un_rot = Lev.angular_momemtum(posR_un_rot,velR_un_rot,MR[unR])
LR_bn_rot, LtR_bn_rot = Lev.angular_momemtum(posR_bn_rot,velR_bn_rot,MR[bnR])
```

```python
print('b = 1')
print('------')
print('The L unbound: Lx = '+str(Lt_un[0])+', Ly = '+str(Lt_un[1])+' and Lz = '+str(Lt_un[2]))
print('The rotated L unbound: Lx = '+str(Lt_un_rot[0])+', Ly = '+str(Lt_un_rot[1])+' and Lz = '+str(Lt_un_rot[2]))
print('------')
print('The L bound: Lx = '+str(Lt_bn[0])+', Ly = '+str(Lt_bn[1])+' and Lz = '+str(Lt_bn[2]))
print('The rotated L bound: Lx = '+str(Lt_bn_rot[0])+', Ly = '+str(Lt_bn_rot[1])+' and Lz = '+str(Lt_bn_rot[2]))
print('-------------------------------------------------------------------------------------------')
print('Renzo')
print('------')
print('The L unbound: Lx = '+str(LtR_un[0])+', Ly = '+str(LtR_un[1])+' and Lz = '+str(LtR_un[2]))
print('The rotated L unbound: Lx = '+str(LtR_un_rot[0])+', Ly = '+str(LtR_un_rot[1])+' and Lz = '+str(LtR_un_rot[2]))
print('------')
print('The L bound: Lx = '+str(LtR_bn[0])+', Ly = '+str(LtR_bn[1])+' and Lz = '+str(LtR_bn[2]))
print('The rotated L bound: Lx = '+str(LtR_bn_rot[0])+', Ly = '+str(LtR_bn_rot[1])+' and Lz = '+str(LtR_bn_rot[2]))
```

```python
fig, axs=plt.subplots(nrows=1,  ncols=2, figsize=(16,8))
fig.subplots_adjust(hspace=0.3)

axs[0].set_title(r'b = 1', fontsize=22)
axs[0].plot(x[np.argmax(d)],z[np.argmax(d)],'.k', label = 'cm particle')
axs[0].arrow(x[np.argmax(d)],z[np.argmax(d)], Lt_bn[0],Lt_bn[2], color='r', label = 'L', width = 0.1,
             head_width = 0.8, head_length=4.5)
axs[0].arrow(x[np.argmax(d)],z[np.argmax(d)], Lt_bn_rot[0],Lt_bn_rot[2], color='b', label = 'Rotated L',
             width = 0.1, head_width = 0.8, head_length=4.5)
axs[0].set_ylabel(r'$x$', fontsize = 20)
axs[0].set_xlabel(r'$z$', fontsize = 20)
axs[0].legend(fontsize = 15, loc=3)

axs[1].set_title(r'Renzo', fontsize=22)
axs[1].plot(xR[np.argmax(d)],zR[np.argmax(d)],'.k', label = 'cm particle')
axs[1].arrow(xR[np.argmax(d)],zR[np.argmax(d)], LtR_bn[0],LtR_bn[2], color='r', label = 'L', width = 0.005,
             head_width = 0.02, head_length=0.02)
axs[1].arrow(xR[np.argmax(d)],zR[np.argmax(d)], LtR_bn_rot[0],LtR_bn_rot[2], color='b', label = 'Rotated L',
            width = 0.002, head_width = 0.01, head_length = 0.03)
axs[1].set_xlabel(r'$z$', fontsize = 20)
axs[1].legend(fontsize = 15)

fig.tight_layout()
plt.show()
```
